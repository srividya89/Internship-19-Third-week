```python
Ratings=[]
Rating=driver.find_elements_by_xpath("//div[@class='_3LWZlK _1BLPMq']")
for a in Rating:
    Ratings.append(a.text)
Ratings
```

```python
Review_summarys=[]
Review_summary=driver.find_elements_by_xpath("//p[@class='_2-N8zT']")
for b in Review_summary:
    Review_summarys.append(b.text)
Review_summarys
```

```python
full_reviews=[]
full_review=driver.find_elements_by_xpath("//div[@class='t-ZTKy']")
for c in full_review:
    full_reviews.append(c.text)
full_reviews
```

```python
Ratingss=[]
Review_summaryss=[]
full_reviewss=[]
```

```python



for j in range(2,12):
    next_btn=driver.find_element_by_xpath("//a[@class='_1LKTO3']")
    next_btn.click()
    time.sleep(1)    
    #driver.get(current_url)
    time.sleep(1)
    Ratii=''
    Rev_sum=''
    full_rev=''
    Ratii=driver.find_elements_by_xpath("//div[@class='_3LWZlK _1BLPMq']")
    for o in Ratii:
        Ratingss.append(o.text)
    time.sleep(1)    
    Rev_sum=driver.find_elements_by_xpath("//p[@class='_2-N8zT']") 
    for p in  Rev_sum:
        Review_summaryss.append(p.text)
    time.sleep(1)    
    full_rev=driver.find_elements_by_xpath("//div[@class='t-ZTKy']")
    for q in full_rev:
        full_reviewss.append(q.text)
    time.sleep(1)    
   
   
      
```
