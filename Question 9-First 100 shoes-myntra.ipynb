```python
url="https://www.myntra.com/shoes"
driver.get(url)
```

```python
try:
    price=driver.find_elements_by_xpath("//label[@class='common-customCheckbox vertical-filters-label']")
    for k in price:
        if k.text=="Rs. 6897 to Rs. 13595(1877)":
            k.click()
    
except StaleElementReferenceException as e:
       time.sleep(1)
    
```

```python
try: 
    color=driver.find_elements_by_xpath("//span[@class='colour-label colour-colorDisplay']")
    for l in color:
        if l.text=="Black (16609)":
             l.click()  
            
except StaleElementReferenceException as e:
       time.sleep(1)
               
```

```python
Brands=[]
Brand=driver.find_elements_by_xpath("//h3[@class='product-brand']")
for a in Brand:
    Brands.append(a.text)
Brands
```

```python
shoe_descs=[]
try:
    shoe_desc=driver.find_elements_by_xpath("//h4[@class='product-product']")
    for b in shoe_desc:
        shoe_descs.append(b.text)
except  NoSuchElementException as e:
        shoe_descs.append("_")
shoe_descs
```

```python
prices=[]
price=driver.find_elements_by_xpath("//div[@class='product-price']")
for c in price:
    prices.append(c.text)
prices
```

```python
Nos=[]
for i in range(1,len(Brands)+1):
    Nos.append(i)
Nos    
```

```python
Brandss=[]
shoe_descss=[]
pricess=[]
```

```python
for j in range(2,4):
    next_btn=driver.find_element_by_xpath("//a[@rel='next']")
    next_btn.click()
    time.sleep(1)    
    #driver.get(current_url)
    time.sleep(1)
    Brande=''
    Product_descse=''
    pricesse=''
    Brande=driver.find_elements_by_xpath("//h3[@class='product-brand']")
    for o in Brande:
        Brandss.append(o.text)
    time.sleep(1) 
    try:
        Product_descse=driver.find_elements_by_xpath("//h4[@class='product-product']") 
        for p in Product_descse:
             shoe_descss.append(p.text)
    except  NoSuchElementException as e:
             shoe_descss.append("_")
    time.sleep(1)    
    pricesse=driver.find_elements_by_xpath("//div[@class='product-price']")
    for q in pricesse:
        pricess.append(q.text)
    time.sleep(1)    
   
```
